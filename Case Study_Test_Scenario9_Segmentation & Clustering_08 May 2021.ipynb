## Since MARITAL_STATUS and AGE NaN is high in propotion to dataset available, droping the rows is not advisable , hence going through the impute approach
## 1) Martial_Status NaN values to be replaced with 'Unknown'
## 2) Age NaN to be replaced with 'Mean' value

```python
dups = df.duplicated()
print('Number of duplicate rows = %d' % (dups.sum()))
print(df.shape)
```

```python
colors=['red','yellow','green','blue']
df.MARITAL_STATUS .value_counts(normalize=True).plot(kind='pie',autopct='%1.0f%%', colors=colors)
```

## Replacing NaN with 'Unknown'  

```python
df['MARITAL_STATUS'].fillna("Unknown", inplace = True)
```

```python
df.head()
```

##  Age NaN to be replaced with 'Mean' value

```python
df['AGE'] = df['AGE'].fillna(df['AGE'].mean())
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# The following code loops through each column and checks if the column type is object then converts those columns
# into categorical with each distinct value becoming a category or code.
for feature in df.columns: 
    if df[feature].dtype == 'object': 
        df[feature] = pd.Categorical(df[feature]).codes
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
HClust = linkage(df, method = 'ward')
```

```python
dend = dendrogram(HClust)
```

## Plot the truncated dendrogram  

```python
dend = dendrogram(HClust,
                 truncate_mode='lastp',
                 p = 25,# we are looking at the last 25 merges
                 )
```

```python
dend = dendrogram(HClust,
                 truncate_mode='lastp',
                 p = 15,# we are looking at the last 15 merges
                 )
```

```python
dend = dendrogram(HClust,
                 truncate_mode='lastp',
                 p = 10,# we are looking at the last 10 merges
                 )
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
#Method 1

clusters_1 = fcluster(HClust,40, criterion='maxclust')
clusters_1
```

```python
# Method 2

clusters_2 = fcluster(HClust, 35, criterion='distance')
clusters_2
```

```python
Hierarchical_clustering_Output_df =df.copy()
Hierarchical_clustering_Output_df ['H_clusters'] = clusters_1
```

```python
Hierarchical_clustering_Output_df.head()
```

## check whether the number of clusters generated by the 'maxclust' criterion is same as the number of clusters generated by the 'distance' criterion.

```python
np.array_equal(clusters_1,clusters_2)
```

```python
wss =[] 
for i in range(1,15):
    KM = KMeans(n_clusters=i)
    KM.fit(df)
    wss.append(KM.inertia_)
wss
```

```python
k_means = KMeans(n_clusters = 6)
k_means.fit(df)
labels = k_means.labels_
labels
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
silhouette_score(df,labels)
```
